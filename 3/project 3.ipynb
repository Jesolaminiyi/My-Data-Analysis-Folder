```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fifa_2022_df = pd.read_csv('c:\\Users\\User\\Desktop\\All Files\\Everything Data\\Analysis Folder\\Python Analysed Dataset\\3\\Project 3\\FIFA - 2022.csv')
```

```python
fifa_2022_df.head()
```

```python
fifa_2022_df.isnull().sum()
```

```python
fifa_2022_df.dropna(how = "all").head()
```

```python
Numbers_of_team = fifa_2022_df["Team"].unique()
len(Numbers_of_team)
```

```python
Average_goal_against = fifa_2022_df["Goals Against"].sum() / len(fifa_2022_df)

round(Average_goal_against,2)
```

```python
Average_goal_scored = fifa_2022_df["Goals For"].sum() / len(fifa_2022_df)

round(Average_goal_scored, 2)
```

```python
Highest_goal_scored = fifa_2022_df["Goals For"].max()

Highest_goal_scored
```

```python
Lowest_Goal_scored = fifa_2022_df["Goals For"].min()

Lowest_Goal_scored
```

```python
Highest_goal_against = fifa_2022_df["Goals Against"].max()

Highest_goal_against
```

```python
Lowest_goal_against = fifa_2022_df["Goals Against"].min()

Lowest_goal_against
```

```python
Goals_ranking = fifa_2022_df.groupby("Team")["Goals For"].sum()
```

```python
Goals_ranking.plot(kind = "bar", figsize = (20,5))
plt.title("2022 Total Goal Scored by Team")
plt.show()
```

```python
Goal_scored_and_Goal_against = fifa_2022_df.pivot_table(index = "Team", values = ["Goals For", "Goals Against"], aggfunc= "sum")
Goal_scored_and_Goal_against.plot(kind = "bar", figsize=(20,5), title = "Goals scored and Goals against by Team" )
plt.show()
```

```python
Ranking_by_point = fifa_2022_df.pivot_table(index = "Team", values = "Points", aggfunc="sum")
```

```python
Ranking_by_point.plot(kind = "bar", figsize = (15,5), title = "RANKING BY POINT")
plt.show()
```

```python
fifa_2022_df["Avg Goal Per Game"] = fifa_2022_df["Goals For"] / fifa_2022_df["Games Played"]
```

```python
fifa_2022_df["Avg Goal Per Game"] = fifa_2022_df["Avg Goal Per Game"].astype(int) 
```

```python
fifa_2022_df.head()
```

```python
pivot  = fifa_2022_df.pivot_table(index = "Team", values = ["Win","Loss","Draw"], aggfunc="sum")

pivot.plot(kind = "bar", figsize= (20,5))
plt.title("Total_GOal_scored and Goal_Against And Draw by Team")
plt.show()
```
